```python
df = ords_prods_merged [:1000000]
```

```python
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
df['price_range'] = df.apply(price_label, axis=1)
```

```python
#Q2 CREATING_LIST

result_1 = []

for value in ords_prods_merged["order_dow"]:
  if value == 0 or  value == 1:
    result_1.append("Busiest days")
  elif value == 4 or value == 3:
    result_1.append("Least busy days")
  else:
    result_1.append("Regularly busy days")
```

```python
#Q4  CREARING_LIST_OF_ORDERS_HOURS

result_2 = []

for value in ords_prods_merged["order_hour_of_day"]:
  if value == 9 or value == 10 or value == 11 or value == 12 or value == 13 or value == 14 or value == 15 or value == 16:
    result_2.append("Most orders")
  elif value == 4 or value == 3 or value == 2 or value == 1 or value == 5:
    result_2.append("Fewest orders")
  else:
    result_2.append("Average orders")
```
